# Challenge Draw a bar plot

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../data/gapminder_total_fertility.csv', index_col = 0)
```

```python
df.columns = df.columns.astype(int)
```

```python
# select 3 years 
df = df[[1950, 1955, 2000]]
```

```python
# select 4 countries 
df = df.loc[['Germany', 'India', 'Bulgaria', 'Kenya']]
```

```python
df.plot.bar(figsize = (5,7))
plt.savefig('bars.png')
```

```python
df.plot.bar(y = 1950, width = 0.7, color = 'orange')
```

```python
df_transpose = df.transpose()
```

```python
df_transpose.plot.bar()
```

# even more penguine stuff 
found this tutorial here [seaborn tutorial](https://seaborn.pydata.org/tutorial/distributions.html)

```python
import pandas as pd
import seaborn as sns
sns.set()
```

```python
penguins = sns.load_dataset('penguins')
```

```python
sns.displot(penguins, x = 'flipper_length_mm')
```

```python
# choose a binsize
sns.displot(penguins, x = 'flipper_length_mm', binwidth = 3)
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.displot(tips, x = 'size')
```

```python
# bins are to small => not a nice graph
# for discrete values one can set discrete to true so the bins will fill the whole 
sns.displot(tips, x = 'size', discrete = True)
```

```python
sns.displot(tips, x = 'day', shrink = .8)
```

# Conditioning on other variables

```python
sns.displot(penguins, x = 'flipper_length_mm', hue = 'species')
```

this is hard to read because the plots are overlapping

```python
sns.displot(penguins, 
            x = 'flipper_length_mm', 
            hue = 'species', 
            multiple = 'stack')
```

```python
sns.displot(penguins, 
            x = 'flipper_length_mm', 
            hue = 'sex', 
            multiple = 'dodge')
```

```python
sns.displot(penguins, 
            x = 'flipper_length_mm', 
            hue = 'species', 
            multiple = 'dodge')
```

```python
sns.displot(penguins, 
            x="flipper_length_mm", 
            col="sex", 
            hue = 'species', 
            multiple="dodge")
```

```python
# normalizing each set individually
sns.displot(penguins, 
            x="flipper_length_mm", 
            hue="species", 
            stat="density", 
            common_norm=False)
```

## Kernel density estimation - kde

```python
sns.displot(penguins, 
            x = 'flipper_length_mm', 
            kind = 'kde')
```

```python
sns.displot(penguins, 
            x = 'flipper_length_mm', 
            hue = 'species', 
            col = 'sex', 
            kind = 'kde')
```

```python
sns.displot(penguins, 
            x = 'flipper_length_mm', 
            hue = 'species', 
            kind = 'kde', 
            multiple = 'stack')
```

## bivariant distributions 

```python
sns.displot(penguins, 
            x = 'bill_length_mm', 
            y = 'bill_depth_mm')
```

```python
sns.displot(penguins, 
            x = 'bill_length_mm', 
            y = 'bill_depth_mm', 
            kind = 'kde')
```

```python
sns.displot(penguins, 
            x = 'bill_length_mm', 
            y = 'bill_depth_mm', 
            hue = 'species')
```

```python
sns.displot(penguins, 
            x = 'bill_length_mm', 
            y = 'bill_depth_mm', 
            hue = 'species', 
            kind = 'kde')
```

```python
sns.displot(penguins, 
            x="bill_length_mm", 
            y="bill_depth_mm", 
            binwidth=(2, .5), 
            cbar=True)
```

## plotting joint and marginal distributions 

```python
sns.jointplot(data = penguins, 
              x = 'bill_length_mm', 
              y = 'bill_depth_mm')
```

```python
sns.jointplot(data = penguins, 
              x = 'bill_length_mm', 
              y = 'bill_depth_mm',
              hue = 'species',
              kind = 'kde')
```

```python
sns.displot(penguins, 
            x = 'bill_length_mm', 
            y = 'bill_depth_mm', 
            kind = 'kde', 
            rug = True)
```

```python
sns.relplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
sns.rugplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

## Plotting many distributions

```python
sns.pairplot(penguins)
```

```python
sns.pairplot(penguins, hue = 'species')
```

```python
sns.pairplot(penguins, hue = 'species', kind = 'kde')
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill = True)
g.map_diag(sns.histplot, kde = True)
```

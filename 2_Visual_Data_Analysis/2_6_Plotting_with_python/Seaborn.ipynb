```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("../data/pokemon.csv",index_col=0)
```

```python
df.head()
```

```python
sns.boxplot(data=df.iloc[:,4:8], orient="h")
plt.savefig('sns_boxplot.png', dpi=150)
```

```python
sns.countplot(x="Type 1",data=df,palette="rainbow")
plt.xticks(rotation=-45)
plt.savefig('sns_countplot.png', dpi=150)
```

```python
sns.kdeplot(df.Attack,df.Defense,shade=False)
sns.scatterplot(df['Attack'],df['Defense'], size=0.03, legend=False)
# plt.savefig('sns_kdeplot.png', dpi=150)
```

```python
sns.heatmap(df.corr(),annot = True)
plt.savefig('sns_heatmap.png', dpi=150)
```

```python
data = df[['Attack', 'Defense', 'Speed', 'HP', 'Legendary']]
sns.pairplot(data, hue='Legendary', diag_kind="kde", kind="scatter", palette="husl")
plt.savefig('sns_pairplot.png', dpi=150)
```

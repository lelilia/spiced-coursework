```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
x_vals = range(1, 101)
y_vals = [2*x for x in x_vals]
```

```python
plt.plot(x_vals, y_vals)
plt.show()
```

```python
# customizing
plt.plot(x_vals, 
         y_vals,
         color = 'green',
         marker = 'x',
         linestyle = 'dashed',
         linewidth = 2,
         markersize = 3
        )
plt.title('What a line')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend(['y = 2x'])
plt.grid()
```

```python
# Plotting multiple Series
y_vals2 = [4*x for x in x_vals]

plt.plot(x_vals, y_vals)
plt.plot(x_vals, y_vals2)
plt.title('Two lines!')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y = 2x', 'y = 4x'])
plt.grid()
```

```python
# in a more object oriented way one could also

fig, ax = plt.subplots() # creates an empty figure and axes

ax.plot(x_vals, y_vals)
ax.plot(x_vals, y_vals2)
ax.set_title('the same in more object oriented')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(['y = 2x', 'y = 4x'])
ax.grid()
```

```python
# multiple plots
y_vals2 = [np.sin(x) for x in x_vals]

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, sharex = True)

ax1.plot(x_vals, y_vals)
ax1.set_title('y = 2x')
ax1.set_ylabel('y')

ax2.plot(x_vals, y_vals2)
ax2.set_title('y = sin(x)')
ax2.set_ylabel('y')
ax2.set_xlabel('x')

plt.tight_layout()
```

```python
# line plot
x = list(range(1,11))
y = [2, 8, 0, 3, 5, 6, 1, 9, 10, 3]

plt.plot(x, y)
plt.show()
```

```python
# scatterplot
n_balls = 50
x = [random.triangular() for i in range(n_balls)]
y = [random.gauss(0.5, 0.15) for i in range(n_balls)]
colors = [random.randint(1,4) for i in range(n_balls)]
areas = [math.pi * random.randint(5,15)**2 for i in range(n_balls)]

plt.scatter(x, y, s = areas, c = colors, alpha = 0.75)
plt.show()
```

```python
# bar chart
y = [213, 931, 13]
x = ['green', 'blue', 'apple']

plt.bar(x, y)
plt.ylabel('jo')
plt.yticks(rotation = -45)
plt.show()
```

```python
# histogram
data = [random.gauss(1000, 20) for i in range(1000)]

plt.hist(data, 30, histtype='bar', 
         stacked=False,  cumulative=False, 
         label=("Hello","Win"),
         facecolor='green', alpha=0.75)
plt.show()
```

```python
# pie chart

labels = 'mate', 'water', 'mate mate', 'tea'
count = [120, 23, 200, 2]
explode = [0.0, 0.0, 0.1, 0.0]

colors = ['gold', 'pink', 'aqua', 'blue']

plt.pie(count, 
        explode= explode, 
        labels=labels, 
        shadow = True, 
        autopct="%1.0F%%",
        colors = colors)
plt.show()
```

```python
# plotting functions

from pylab import *

x = [0.01 * i for i in range(1000)]
y = [math.sin(j) for j in x]

plot(x, y, 'k-', linewidth = 1)
text(5, 0, "$y = sin(x)$", horizontalalignment='center', fontsize = 20)
axis([0,3 * math.pi, -1.2, 1.2])
```

```python
# subplots

fig, ax = plt.subplots(3,1,figsize=(10,3))
num = 1000

# size of spots
s = 121

x1 = np.linspace(-0.5,1,num) + (0.5 - np.random.rand(num))
y1 = np.linspace(-5,5,num) + (0.5 - np.random.rand(num))
x2 = np.linspace(-0.5,1,num) + (0.5 - np.random.rand(num))
y2 = np.linspace(5,-5,num) + (0.5 - np.random.rand(num))
x3 = np.linspace(-0.5,1,num) + (0.5 - np.random.rand(num))
y3 = (0.5 - np.random.rand(num))

ax[0].scatter(x1, y1, color='r', s=2*s, marker='^', alpha=.4)
ax[1].scatter(x2, y2, color='b', s=s/2, alpha=.4)
ax[2].scatter(x3, y3, color='g', s=s/3, marker='s', alpha=.4)
ax[0].set_title("Example of a Scatter plot which is divided into subplots")
```

```python
# 3d plot

from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection = '3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride =1, cstride = 5, cmap = cm.coolwarm)
plt.show()
```
